# Исследование объявлений о продаже квартир

## 1 Цели исследования

<p style="text-align:justify;">Цель проводимого исследования &mdash; определение рыночной стоимости объектов недвижимости и установление параметров, позволяющих построить автоматизированную систему для отслеживания аномалий и мошеннической деятельности.</p>

## 2 Описание данных

<p style="text-align:justify;">Исходными данными настоящего исследования является архив объявлений на Яндекс.Недвижимость о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.</p>

Таблица исходных данных содержит следующие столбцы:
- airports_nearest — расстояние до ближайшего аэропорта в метрах (м);
- balcony — число балконов;
- ceiling_height — высота потолков (м);
- cityCenters_nearest — расстояние до центра города (м);
- days_exposition — сколько дней было размещено объявление (от публикации до снятия);
- first_day_exposition — дата публикации;
- floor — этаж;
- floors_total — всего этажей в доме;
- is_apartment — апартаменты (булев тип);
- kitchen_area — площадь кухни в квадратных метрах (м²);
- last_price — цена на момент снятия с публикации;
- living_area — жилая площадь в квадратных метрах(м²);
- locality_name — название населённого пункта;
- open_plan — свободная планировка (булев тип);
- parks_around3000 — число парков в радиусе 3 км;
- parks_nearest — расстояние до ближайшего парка (м);
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м);
- rooms — число комнат;
- studio — квартира-студия (булев тип);
- total_area — площадь квартиры в квадратных метрах (м²);
- total_images — число фотографий квартиры в объявлении.

### 3.1 Глобальные переменные и функции

<p style="text-align:justify;">Объявление глобальных переменных и функций.</p>

```python
def ticks_millions_rnd0(x, pos):
    """ 
    Функция форматирования меток осей с округлением до миллиона
    """
    return '{:.0f}'.format(round(x * 1e-6))

def ticks_millions_rnd2(x, pos):
    """ 
    Функция форматирования меток осей с округлением до сотых миллиона
    """
    return '{:0.2f}'.format(round(x * 1e-6, 2))

def ticks_ones_rnd0(x, pos):
    """ 
    Функция форматирования меток осей с округлением до целого
    """
    return '{:.0f}'.format(round(x))
```

<p style="text-align:justify;">В пункте <a href="https://yandex.ru/support/realty/rules/requirements-ads.html">"Требования к объявлениям"</a> настоящих (на 02.06.2021) правил размещения объявлений на Яндекс.Недвижимость устанавливается минимальное количество прикрепляемых к объявлениям фотографий.</p> 
<ui>

<p style="text-align:justify;">Таким образом, минимальное количество фотографий для жилых объектов &mdash; 3, для нежилых &mdash; 2. В исходной же таблице минимальное значение &mdash; 0. Это противоречит правилам сервиса, если предположить, что в период сбора статистики данные правила совпадали с настоящими.</p>

### 5.3 Категоризация этажа квартиры: первый, последний, другой

<p style="text-align:justify;">Введем три категории: First (первый), Last (последний) и Other (другой). Определим, к какой категории относится этаж продаваемой квартиры. Результат сохраним в столбец <b>floor_category</b>. Очевидно, что для определения, является ли этаж первым достаточно значений из столбца <b>floor</b>. В то же время для определения, является ли этаж последним или другим (промежуточным), понадобятся значения из столбца <b>floors_total</b>, который, в отличие от <b>floor</b>, содержит пропуски.</p>

<p style="text-align:justify;">Результирующий столбец <b>floor_category</b> будет иметь пропуски только в тех случаях, когда <b>floors_total</b> содержит пропуск и при этом значение <b>floor</b> не равно 1.</p>

```python
def floor_category(row):
    if row.floor == 1:
        return 'First'
    elif not pd.isna(row.floors_total):
        if row.floor == row.floors_total:
            return 'Last'
        else:
            return 'Other'
    else:
        return pd.NA
data['floor_category'] = data.apply(floor_category, axis=1)
data.floor_category.head(10)
```

#### 6.3.1 Площадь

<p style="text-align:justify;">Для изучения зависимости цены квартиры от площади, построим для обеих величин сводную таблицу, диаграмму расеяния, а также вычислим линейный коэффициент корреляции.<p>
    
<p style="text-align:justify;">Так как площадь квартиры &mdash; это количественная характеристика, построим сводную таблицу для ее интервалов, взятых с шагом 50 м<sup>2</sup>. В каждом интервале вычислим количество объявлений, среднюю и медианную цены на квартиру.<p>

```python
# Шаг
total_area_step = 50

# Значения, округленные снизу до ближайших кратных интервалу
total_area_steps_a = total_area_step * (data.total_area / total_area_step).astype(int)

# Значения, округленные сверху до ближайших кратных интервалу
total_area_steps_b = total_area_step * (1 + data.total_area / total_area_step).astype(int)

# По этому столбцу будет индексами сводной таблицы
data['total_area_stepped_50sqm'] = total_area_steps_b

# Сводная таблица
data_price_area_pivot = data.pivot_table(
    index='total_area_stepped_50sqm',
    values='last_price',
    aggfunc=['count', 'mean', 'median'])

cn1 = 'Объявлений, шт.'
cn2 = 'Средняя цена, ' + html.unescape('&#8381;')
cn3 = 'Медианная цена, ' + html.unescape('&#8381;')
data_price_area_pivot.columns = [cn1, cn2, cn3]

data_price_area_pivot.index = pd.Series(total_area_steps_a.sort_values().unique().astype(str))\
+ '-' + pd.Series(total_area_steps_b.sort_values().unique().astype(str))
data_price_area_pivot.index.rename('Площадь, м\u00b2', inplace=True)

data_price_area_pivot[cn2] = data_price_area_pivot[cn2].apply(round)
data_price_area_pivot[cn3] = data_price_area_pivot[cn3].apply(round)

data_price_area_pivot.transpose()
```

<p style="text-align:justify;">Отобразим сводную таблицу в виде столбчатых диаграмм.</p>

```python
# График и оси
fig, ax = plt.subplots(1, 2, figsize=(14, 3))
ax1, ax2 = ax[0], ax[1]

# Общее название группы графиков
fig.suptitle('Зависимость цены квартиры от площади', fontsize=14, y=1.025)

# Гистограмма
ax1.bar(
    data_price_area_pivot.index,
    data_price_area_pivot[cn2]
)
ax1.set_title('Среднее значение')
ax1.set_xlabel('Площадь, м\u00b2')
ax1.set_ylabel('Цена, млн ' + html.unescape('&#8381;'))
ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey', alpha=0.5)
ax1.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(ticks_millions_rnd0))
ax1.tick_params(axis='x', labelrotation=30)

# Гистограмма
ax2.bar(
    data_price_area_pivot.index,
    data_price_area_pivot[cn3]
)
ax2.set_title('Медиана')
ax2.set_xlabel('Площадь, м\u00b2')
ax2.set_ylabel('Цена, млн ' + html.unescape('&#8381;'))
ax2.yaxis.grid(True, linestyle='-', which='major', color='lightgrey', alpha=0.5)
ax2.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(ticks_millions_rnd0))
ax2.tick_params(axis='x', labelrotation=30)

fig.tight_layout()
plt.show()
```

<p style="text-align:justify;">Для лучшего визуального восприятия диаграммы рассеяния будем использовать логарифмическую шкалу палитры.</p>

```python
# График и оси
fig, ax = plt.subplots(figsize=(10, 8))

# Общее название группы графиков
fig.suptitle('Зависимость цены квартиры от площади', fontsize=14, y=1.025)

# Гистограмма
ax.hexbin(
    data.total_area,
    data.last_price,
    gridsize=50,
    bins='log',
    cmap='inferno'
)
ax.set_xlabel('Площадь, м\u00b2')
ax.set_ylabel('Цена, млн ' + html.unescape('&#8381;'))
ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(ticks_millions_rnd0))

fig.tight_layout()
plt.show()
```

<p style="text-align:justify;">Линейный коэффициент корреляции Пирсона.</p>

<p style="text-align:justify;">Зависимости средней и медианной цен квартиры от ее площади повторяют друг друга. <p style="text-align:justify;">Для более точного определения границы центра Санкт-Петербурга округлим расстояния от него до квартир в столбце <b>cityCenters_nearest</b> до целых значений.</p>

```python
data_spb = data[data.locality_name == 'Санкт-Петербург']
```

```python
# Сводная таблица
data_spb_price_centers_nearest_pivot = data_spb.pivot_table(
    index='centers_nearest_stepped_1km',
    values='last_price',
    aggfunc=['count', 'mean'])

cn1 = 'Обявлений, шт'
cn2 = 'Средняя цена, ' + html.unescape('&#8381;')

data_spb_price_centers_nearest_pivot.columns = [cn1, cn2]
data_spb_price_centers_nearest_pivot.index.rename('Дистанция, км', inplace=True)

data_spb_price_centers_nearest_pivot[cn2] =\
    data_spb_price_centers_nearest_pivot[cn2].apply(round)
```

<p style="text-align:justify;">Сводная таблица получилась достаточно громоздкой. Не будем ее выводить, а сразу отобразим в виде столбцовой диаграммы. Ограничимся средним значением цены объекта недвижимости.</p>
