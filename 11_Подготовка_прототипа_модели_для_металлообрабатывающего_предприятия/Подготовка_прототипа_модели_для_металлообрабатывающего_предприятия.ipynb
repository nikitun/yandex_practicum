```python
def df_info(data_frame, nulls_percentage_round=1, deep_memory_usage=False):
    '''
    Выводит на экран общую информацию о таблице
    Входные параметры:
        data_frame             — таблица типа pd.DataFrame;
        nulls_percentage_round — знак, до которого происходит округление в столбце nulls_percentage отображаемой таблицы;
        deep_memory_usage      — глубокий анализ используемой памяти
    '''
    assert isinstance(data_frame, pd.DataFrame), 'Аргумент data_frame должен иметь тип pd.DataFrame'
    t = pd.concat(
        [
         
            data_frame.isna().sum().to_frame('nulls_count'),
            (100 * data_frame.isna().sum() / len(data_frame)).round(nulls_percentage_round).to_frame('nulls_percentage'),
            data_frame.dtypes.to_frame('dtype')
        ],
        axis=1
    )
    s = 'dtypes:'
    c = data_frame.dtypes.value_counts()
    for i in c.index:
        s += ' {0}({1})'.format(i, c[i])
    m = data_frame.memory_usage(deep=deep_memory_usage).sum()
    if int(m / 1024 ** 3):
        d, b = 1024 ** 3, 'GB'
    elif int(m / 1024 ** 2):
        d, b = 1024 ** 2, 'MB'
    else:
        d, b = 1024, 'KB'
    print(t)
    print('columns: ', data_frame.shape[1], ', rows: ', data_frame.shape[0], sep='')
    print(s)
    print('memory_usage: ', round(m / d, 1), '+ ' if not deep_memory_usage else ' ', b, sep='')
    
def recovery(f, c, t):
    '''
    Возвращает эффективность обогащения руды в процентах.
    Входные параметры:
        f — доля золота в процентах в сырье/концентрате до флотации/очистки;
        c — доля золота в процентах в концентрате после флотации/очистки;
        t — доля золота в процентах в отвальных хвостах после флотации/очистки
    '''
    assert isinstance(f, pd.Series) and isinstance(f, pd.Series) and isinstance(f, pd.Series),\
        'Аргументы должны иметь тип pd.Series'
    return 100 * (c * (f - t)) / (f * (c - t))

def smape(y, yp):
    '''
    Возвращает sMAPE — симметричное среднее абсолютное процентное отклонение
    Входные параметры:
        y — вектор реальных значений целевого признака;
        yp — вектор оценок значений целевого признака
    '''
    return 100 / len(y) * np.sum(2 * np.abs(yp - y) / (np.abs(y) + np.abs(yp)))    

def smape_resulted(rf, rfp, **kwargs):
    '''
    Возвращает итоговый sMAPE — метрику оценки качества предсказания
    Входные параметры:
        rf — матрица реальных значений двумерного целевого признака;
        rfp — матрица оценок значений двумерного целевого признака
    '''
    return 0.25 * smape(rf[:,0], rfp[:,0]) + 0.75 * smape(rf[:,1], rfp[:,1])
```

### 3.2 Сравнение размеров гранул сырья на обучающей и тестовой выборках

Сравним размер гранул сырья для обучающей и тестовой выборок. Наличие существенных различий в их распределениях может привести к неправильным оценкам модели.

Будем анализировать размер гранул чернового концентрата `rougher.input.feed_size`, подаваемого на вход флотационной установки, т.к. размер гранул на следующих этапах очистки будет зависеть от него. Построим гистограммы и выведем описательную статистику для данных величин.

```python
pd.concat(
    [d['rougher.input.feed_size'].describe().to_frame(d.name) for d in [data_train, data_test]],
    axis=1
)
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
s1, s2 = data_train['rougher.input.feed_size'], data_test['rougher.input.feed_size']
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(16, 6))
sns.histplot(ax=axs, data=s1, bins=50, color=next(palette), kde=True)
sns.histplot(ax=axs, data=s2, bins=50, color=next(palette), kde=True)
axs.legend(labels=['data_train', 'data_test'], fontsize=12);
fig.suptitle('Распределение размеров гранул сырья в обучающей и тестовой выборках', fontsize=16)
plt.show();
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
s1, s2 = data_train['rougher.input.feed_size'], data_test['rougher.input.feed_size']
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(16, 6))
sns.histplot(
    ax=axs, data=s1[(s1 >= s1.quantile(.025)) & (s1 <= s1.quantile(.975))], bins=30, color=next(palette), kde=True
)
sns.histplot(
    ax=axs, data=s2[(s2 >= s2.quantile(.025)) & (s2 <= s2.quantile(.975))], bins=30, color=next(palette), kde=True
)
axs.legend(labels=['data_train', 'data_test'], fontsize=12);
fig.suptitle(
    'Распределение размеров гранул сырья в обучающей и тестовой выборках (отброшены 5% экстремальных значений)',
    fontsize=16
)
plt.show();
```

Размер гранул сырья в обучающей и тестовой выборках имеет похожий диапазон, в обеих выборках имеют место выбросы. При этом формы распределений заметно различаются. В случае обучающей выборки распределение напоминает нормальное, вытянутое в сторону больших значений. В случае тестовой &mdash; форма распределения далека от нормальной, но она также вытянута в сторону больших значений, по огибающей кривой можно заметить наличие двух мод. В обучающей выборке наиболее часто встречающиеся размеры лежат в диапазоне 45-60 единиц, в то время как в тестовой выборке в этом диапазоне на гистограмме наблюдается провал.

Проверим обе выборки на принадлежность одной генеральной совокупности с помощью парного T-теста для критического уровня статистической значимости 5%, для этого сформулируем две гипотезы:
- H0: значения размеров гранул сырья в обучающей и тестовой выборках принадлежат одной и той же генеральной совокупности;
- H1: не принадлежат.

```python
alpha = 0.05
_, p = st.ttest_ind(
    s1.dropna(),
    s2.dropna()
)
print('p =', p)
if p > alpha:
    print('Выборки принадлежат одной генеральной совокупности (H0 отвергнуть не удалось)')
else:
    print('Выборки принадлежат разным генеральным совокупностям (H0 отвергается)')
```

Отбросим 2% экстремальных значений и повторим проверку гипотезы.

```python
_, p = st.ttest_ind(
    s1[(s1 >= s1.quantile(.01)) & (s1 <= s1.quantile(.99))].dropna(),
    s2[(s2 >= s2.quantile(.01)) & (s2 <= s2.quantile(.99))].dropna()
)
print('p =', p)
if p > alpha:
    print('Выборки принадлежат одной генеральной совокупности (H0 отвергнуть не удалось)')
else:
    print('Выборки принадлежат разным генеральным совокупностям (H0 отвергается)')
del s1, s2
```

В обоих случаях H0 отвергается с вероятностью ошибки первого рода **p** близкой к 0. С другой стороны средние значения размеров гранул отличаются незначительно и составляют 58.41 и 55.95 соответственно. Стандартные отклонения тоже имеют близкие значения: 20.44 и 19.08. Понадеемся, что такие различия не станут причиной большого количества ошибок предсказания модели.

```python
# Черновой концентрат
rougher_inputs_sum =\
data_full['rougher.input.feed_ag']\
+ data_full['rougher.input.feed_pb']\
+ data_full['rougher.input.feed_sol']\
+ data_full['rougher.input.feed_au'];
# Концентрат после флотации
rougher_outputs_sum =\
data_full['rougher.output.concentrate_ag']\
+ data_full['rougher.output.concentrate_pb']\
+ data_full['rougher.output.concentrate_sol']\
+ data_full['rougher.output.concentrate_au'];
# Концентрат после первой очистки
primary_cleaner_outputs_sum =\
data_full['primary_cleaner.output.concentrate_ag']\
+ data_full['primary_cleaner.output.concentrate_pb']\
+ data_full['primary_cleaner.output.concentrate_sol']\
+ data_full['primary_cleaner.output.concentrate_au'];
# Концентрат после второй очистки (финальный концентрат)
final_outputs_sum =\
data_full['final.output.concentrate_ag']\
+ data_full['final.output.concentrate_pb']\
+ data_full['final.output.concentrate_sol']\
+ data_full['final.output.concentrate_au'];
# Визуализация
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(16, 8))
sns.histplot(ax=axs[0, 0], data=rougher_inputs_sum, bins=100, color=next(palette), kde=True);
sns.histplot(ax=axs[1, 0], data=rougher_outputs_sum, bins=100, color=next(palette), kde=True);
sns.histplot(ax=axs[0, 1], data=primary_cleaner_outputs_sum, bins=100, color=next(palette), kde=True);
sns.histplot(ax=axs[1, 1], data=final_outputs_sum, bins=100, color=next(palette), kde=True);
axs[0, 0].set_title('Черновой концентрат', fontsize=12)
axs[0, 1].set_title('После флотации', fontsize=12)
axs[1, 0].set_title('После первичной очистки', fontsize=12)
axs[1, 1].set_title('После вторичной (финальной) очистки', fontsize=12)
fig.suptitle('Распределение суммы концентраций веществ в концентрате на разных этапах', fontsize=16)
plt.show();
```

Нулевая суммарная концентрация веществ в черновом концентрате **rougher_inputs_sum** могла бы объяснить нулевые значения после очистки, но такие значения не обнаружились. - проверена правильность вычисления эффективности обогащения руды `ougher.output.recovery` в обучающей выборке, для чего вычислена средняя абсолютная ошибка (**MAE**) между полученными и загруженными значениями. Ошибка оказалась близкой к нулю ~1.1e-14;


- проанализировано различие в признаках между обучающей и тестовой выборками. В связи с тем, что некоторые признаки замеряются и/или вычисляются значительно позже остальных, принято решение оставить в обучающей выборке только признаки, присутствующие в тестовой выборке, т.к. она больше всего приближена к реальным данным;


- из обучающей выборки удалены ~10.4% строк, содержащих пропуски, так как не все алгоритмы обучения умеют их корректно обрабатывать. Тестовая выборка содержала ~1.2% строк с пропусками, они были оставлены по той же причине, что и в предыдущем пункте;


- проанализированы концентрации металлов **Au**, **Ag** и **Pb** в концентрате на разных этапах очистки. Средняя концентрация золота линейно росла с каждым этапом, достигая ~44.1% в финальном продукте, при этом концентрации серебра и свинца находились в пределах 3.5%-12% (более подробно п. 3.1);


- исследовано распределение размеров гранул сырья в обучающей и тестовой выборках. Выяснилось, что значения имеют близкие диапазоны, но разные формы распределения. Статистическую гипотезу о принадлежности данных распределений одной генеральной совокупности (парный t-тест) при критическом уровне статистической значимости 95% принять не удалось;


- исследована суммарная концентрация металлов и солей в продукте на разных стадиях обогащения. Полученные значения, равные 0, при неравенстве 0 суммы концентраций данных веществ в черновом концентрате, были удалены из выборок как ошибки;


- признаки в обучающей выборке проверены на наличие высокой корреляции (модуль коэффициенты линейной корреляции Пирсона выше 0.95). Из обучающей и тестовой выборок были удалены 15 признаков, содержащих информацию о концентрации флотационных реагентов и уровне жидкости на этапах флотации и первичной очистки;


- оставшиеся признаки были масштабированы.


На основе подготовленных данных были обучены две модели:
- регрессионный случайный лес;
- линейная регрессия.

Целевым признаком являлся двухэлементный вектор, содержащий эффективность обогащения чернового концентрата `rougher.output.recovery` и эффективность обогащения финального концентрата `final.output.recovery`.

Модели обучались для разных наборов гиперпараметров и с использованием кросс-валидации, разбивающей входную выборку на обучающую и валидационную в соотношении 4:1. По результатам обучения была выбрана лучшая модели. В качестве метрики качества предсказания использовалось итоговое **sMAPE**. Лучший случайный лес и лучшая линейная регрессия сравнивались с примитивным регрессором со стратегией **mean**.

Лучшие результаты показала модель случайного регрессионного леса с параметрами:
- criterion = mse;
- max_depth = 50;
- n_estimators = 10;
- random_state = 1234.

Достигнутое значение итогового **sMAPE** составило 9.11, против 7.64 у примитивного регрессора.
